```python
import numpy as np
import matplotlib.pyplot as plt

# Assume df_train is your DataFrame with columns 'SibSp' and 'Parch'
# Define the full range of categories (0 to 7)
categories = np.arange(8)

# Get counts for each category, ensuring all values 0-7 are present
sibsp_counts = df_train['SibSp'].value_counts().reindex(categories, fill_value=0).sort_index()
parch_counts = df_train['Parch'].value_counts().reindex(categories, fill_value=0).sort_index()

# Number of categories
N = len(categories)

# Compute the angle for each category
angles = np.linspace(0, 2 * np.pi, N, endpoint=False)

# Set bar width as a fraction of the available space
bar_width = (2 * np.pi / N) * 0.4

# Create a polar subplot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})

# Define an offset so the two bars for each category don't overlap
offset = bar_width / 2

# Plot bars for SibSp (shifted slightly to the left)
bars_sibsp = ax.bar(angles - offset, sibsp_counts, width=bar_width,
                    color='blue', alpha=0.6, label='SibSp')

# Plot bars for Parch (shifted slightly to the right)
bars_parch = ax.bar(angles + offset, parch_counts, width=bar_width,
                    color='red', alpha=0.6, label='Parch')

# Set the category labels at the corresponding angles
ax.set_xticks(angles)
ax.set_xticklabels(categories)

# Optional: add a legend
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()
```

#### Possible models: `Logistic Regression` , `Decision Tree` , `Random Forest`


```python
# Features (X) and target (y) for training
X_train = df_train.drop('Survived', axis=1)  # Features for training
y_train = df_train['Survived']  # Target for training

# Features for testing (no target column in df_test)
X_test = df_test  # Use all columns in df_test as features
```

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize the Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)

# Train the model on the training data
rf_model.fit(X_train, y_train)
```
